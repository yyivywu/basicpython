# 圖形元素

線 (line), 點 (marker), 文字 (text), 圖例 (legend), 網格 (grid), 標題 (title), 圖片 (image) 等

* 畫線，用 plt.plot() 或 ax.plot()
* 記號，用 plt.scatter() 或 ax.scatter()
* 文字，用 plt.text() 或 ax.text()
* 圖例，用 plt.legend() 或 ax.legend()
* 圖片，用 plt.imshow() 或 ax.imshow()

```python
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

# 定義顏色碼

```python
r_hex = '#dc2624'     # red,       RGB = 220,38,36
dt_hex = '#2b4750'    # dark teal, RGB = 43,71,80
tl_hex = '#45a0a2'    # teal,      RGB = 69,160,162
r1_hex = '#e87a59'    # red,       RGB = 232,122,89
tl1_hex = '#7dcaa9'   # teal,      RGB = 125,202,169
g_hex = '#649E7D'     # green,     RGB = 100,158,125
o_hex = '#dc8018'     # orange,    RGB = 220,128,24
tn_hex = '#C89F91'    # tan,       RGB = 200,159,145
g50_hex = '#6c6d6c'   # grey-50,   RGB = 108,109,108
bg_hex = '#4f6268'    # blue grey, RGB = 79,98,104
g25_hex = '#c7cccf'   # grey-25,   RGB = 199,204,207
```

```python
# index_col = 0 是說把第一列 Date 當成行標籤 (index)
# parse_dates = True 是說把行標籤轉成 date 物件
# dayFirst = True 是說日期是 DD/MM/YYYY 這樣的格式

data = pd.read_csv('SnP500Index.csv', index_col=0, parse_dates=True, dayfirst=False)
data.head(3).append(data.tail(3))
```

```python
# 讀取某日期區間資料

spx = data[['Adj Close']].loc['2007-01-01':'2010-01-01']

spx.head(3).append(spx.tail(3))
```

```python
# spx 是個 DataFrame，將它的值畫出折線圖只需用 plt.plot() 函式
# plot() 函式只有變數 y 時 (y = spx.values)，自變數預設賦值為 range(len(y))
# 所有設定都用的是 matplotlib 的預設設定

plt.plot(spx.values)
plt.show()
```

```python
# plt.rcParams 可檢視圖形所有預設屬性  (非常多的屬性值)

plt.rcParams
```

```python
# 觀看下列圖表預設屬性值
# 圖表尺寸 (figsize)，每英寸畫素點 (dpi)，橫縱軸刻度 (xticks, yticks)，橫縱軸邊界 (xlim, ylim)
# 線條顏色 (color)，線條風格 (linestyle)，線條寬度 (linewidth)

print('figure size:', plt.rcParams['figure.figsize'])
print('figure dpi:',plt.rcParams['figure.dpi'])
print('line color:',plt.rcParams['lines.color'])
print('line style:',plt.rcParams['lines.linestyle'])
print('line width:',plt.rcParams['lines.linewidth'])

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(spx['Adj Close'].values)

print('xticks:', ax.get_xticks())
print('yticks:', ax.get_yticks())
print('xlim:', ax.get_xlim())
print('ylim:', ax.get_ylim())
```

```python
# 修改圖形屬性值

# figsize(w,h) 決定圖的寬和高 (單位是英寸)
# dpi 全稱 dots per inches，測量每英寸多少畫素
# 兩個屬性一起用，那麼得到的圖的畫素為 (w*dpi, h*dpi)
# figsize 和 dpi 控制圖的大小和畫素
fig = plt.figure(figsize=(16,6), dpi=100)
ax = fig.add_subplot(1, 1, 1)

x = spx.index  # 日期
y = spx.values

# 用 color，linewidth 和 linestyle 屬性控制折線的顏色
ax.plot(x, y, color=dt_hex, linewidth=2, linestyle='-')

#ax.set_xlim(['2007/1/1', '2010/1/1'])
ax.set_ylim(y.min()*0.8, y.max()*1.2)

plt.show()
```

```python
fig = plt.figure(figsize=(16,6), dpi=100)
ax = fig.add_subplot(1, 1, 1)

x = spx.index  # 日期
y = spx.values

# 用 color，linewidth 和 linestyle 屬性控制折線的顏色
ax.plot(y, color=dt_hex, linewidth=2, linestyle='-', label='S&P 500')
ax.legend(loc=0, frameon=True)  # 顯示圖例

# 設定橫軸的邊界，下界是 - 1，上界是 len(x) +1
ax.set_xlim(-1, len(x)+1)
ax.set_ylim(y.min()*0.8, y.max()*1.2)

# 設定橫軸「數值刻度」 -> 0, 40, 80...
ax.set_xticks(range(0, len(x), 40))
ax.set_xticklabels([x[i].strftime('%Y-%m-%d') for i in ax.get_xticks()], rotation=90)

plt.show()
```

```python
# Vix

data = pd.read_csv('vix-2018.csv', index_col=0, parse_dates=True, dayfirst=True)
vix = data[['VIXClose']].loc['2007-01-01':'2010-01-01']

vix.head(3).append(vix.tail(3))
```

```python
# 同一座標系畫兩條線

plt.figure(figsize=(16,6), dpi=100)

x = spx.index
y1 = spx.values
y2 = vix.values

plt.plot(y1, color=dt_hex, linewidth=2, linestyle='-', label='S&P500')
plt.plot(y2, color=r_hex, linewidth=2, linestyle='-', label='VIX')
plt.legend(loc=0, frameon=True)

plt.xlim(-1, len(x)+1)
plt.ylim(np.vstack([y1,y2]).min()*0.8, np.vstack([y1,y2]).max()*1.2)

x_tick = range(0, len(x), 40)
x_label = [x[i].strftime('%Y-%m-%d') for i in x_tick]
plt.xticks(x_tick, x_label, rotation=90)

plt.show()
```

# 兩個座標系 (two axes)

```python
# 用 ax1 和 ax2 實現在兩個座標系上畫圖

fig = plt.figure(figsize=(16,6), dpi=100)
ax1 = fig.add_subplot(1, 1, 1)

x = spx.index  # 日期
y1 = spx.values
y2 = vix.values

ax1.plot(y1, color=dt_hex, linewidth=2, linestyle='-', label='S&P 500')
ax1.set_xlim(-1, len(x)+1)
ax1.set_ylim(np.vstack([y1,y2]).min()*0.8, np.vstack([y1,y2]).max()*1.2)

x_tick = range(0, len(x), 40)
x_label = [x[i].strftime('%Y-%m-%d') for i in x_tick]
ax1.set_xticks(x_tick)
ax1.set_xticklabels(x_label, rotation=90)
ax1.legend(loc='upper left', frameon=True)  

# 設定第二個座標軸
ax2 = ax1.twinx()
ax2.plot(y2, color=r_hex, linewidth=2, linestyle='-', label='VIX')
ax2.legend(loc='upper right', frameon=True)

plt.show()
```

# 兩幅子圖 (two subplots)

```python
# 定義 subplot(2,1,1) 和 subplot(2,1,2) 實現在兩幅子圖上畫圖

plt.figure(figsize=(16,12), dpi=100)

# subplot 1
plt.subplot(2, 1, 1)
x = spx.index
y1 = spx.values

plt.plot(y1, color=dt_hex, linewidth=2, linestyle='-', label='S&P500')
plt.legend(loc='upper left', frameon=True)  

plt.xlim(-1, len(x)+1)
plt.ylim(y1.min()*0.8, y1.max()*1.2)
x_tick = range(0, len(x), 40)
x_label = [x[i].strftime('%Y-%m-%d') for i in x_tick]
plt.xticks(x_tick, x_label, rotation=45)
# plt.xticks([])   # 不要顯示 x 軸座標

# subplot 2
plt.subplot(2, 1, 2)
y2 = vix.values

plt.plot(y2, color=r_hex, linewidth=2, linestyle='-', label='VIX')
plt.xlim(-1, len(x)+1)
plt.ylim(y2.min()*0.8, y2.max()*1.2)

plt.xticks(x_tick, x_label, rotation=45)
plt.legend(loc='upper left', frameon=True)  

plt.show()
```

# 設定標註

在金融危機時期，市場發生了 5 件大事，分別是：
* 2007-10-11: 牛市頂點
* 2008-03-12: 貝爾斯登倒閉
* 2008-09-15: 雷曼兄弟倒閉
* 2009-01-20: 蘇格蘭皇家銀行股票拋售
* 2009-04-02: G20 峰會

```python
fig = plt.figure(figsize=(16,6), dpi=100)

crisis_data = [(datetime(2007,10,11), 'Peak of bull market'), 
               (datetime(2008,3,12), 'ear Stearns Fails'), 
               (datetime(2008,9,15), 'Lehman Bankruptcy'), 
               (datetime(2009,1,20), 'RBS Sell-off'), 
               (datetime(2009,4,2), 'G20 Summit')]

ax1 = fig.add_subplot(1, 1, 1)

x = spx.index  
y1 = spx.values
y2 = vix.values

ax1.plot(y1, color=dt_hex, linewidth=2, linestyle='-', label='S&P 500')
ax1.legend(loc='upper left', frameon=True)

ax1.set_xlim(-1, len(x)+1)
ax1.set_ylim(np.vstack([y1,y2]).min()*0.8, np.vstack([y1,y2]).max()*1.2)

x_tick = range(0, len(x), 40)
x_label = [x[i].strftime('%Y-%m-%d') for i in x_tick]
ax1.set_xticks(x_tick)
ax1.set_xticklabels(x_label, rotation=90)

# 用 for 迴圈讀取 crisis_data 裡面每個日期 date 和事件 label
for date, label in crisis_data:
    date = date.strftime('%Y-%m-%d')
    xi = x.get_loc(date)  # 獲取每一個 date 在整個日期陣列中的索引 xi，以及對應的 spx 值 yi
    yi = spx.asof(date)
    ax1.scatter(xi, yi, 80, color=r_hex)  # 用 scatter() 函式畫出一個圓點，標註事件在 spx 的位置
    # 在 annotate() 函式裡設定了事件，箭頭座標，事件列印的座標，箭頭性質，以及對齊屬性
    ax1.annotate(label, xy=(xi, yi+60), xytext=(xi, yi+300), 
                 arrowprops=dict(facecolor='black', headwidth=4, width=1, headlength=6), 
                 horizontalalignment='left', verticalalignment='top')

# 設定第二個座標軸
ax2 = ax1.twinx()
ax2.plot(y2, color=r_hex, linewidth=2, linestyle='-', label='VIX', alpha=0.3)
ax2.legend(loc='upper right', frameon=True)

plt.show()
```

```python
# 加上關鍵事件的日期

fig = plt.figure(figsize=(16,6), dpi=100)

crisis_data = [(datetime(2007,10,11), 'Peak of bull market'), 
               (datetime(2008,3,12), 'ear Stearns Fails'), 
               (datetime(2008,9,15), 'Lehman Bankruptcy'), 
               (datetime(2009,1,20), 'RBS Sell-off'), 
               (datetime(2009,4,2), 'G20 Summit')]

ax1 = fig.add_subplot(1, 1, 1)

x = spx.index  
y1 = spx.values
y2 = vix.values

ax1.plot(y1, color=dt_hex, linewidth=2, linestyle='-', label='S&P 500')
ax1.legend(loc='upper left', frameon=True)

ax1.set_xlim(-1, len(x)+1)
ax1.set_ylim(np.vstack([y1,y2]).min()*0.8, np.vstack([y1,y2]).max()*1.2)

x_tick = range(0, len(x), 40)
x_label = [x[i].strftime('%Y-%m-%d') for i in x_tick]
ax1.set_xticks(x_tick)
ax1.set_xticklabels(x_label, rotation=90)

init_tick = list(range(0, len(x), 40))
impt_tick = []
impt_date = []

# 用 for 迴圈讀取 crisis_data 裡面每個日期 date 和事件 label
for date, label in crisis_data:
    date = date.strftime('%Y-%m-%d')
    impt_date.append(date)
    
    xi = x.get_loc(date)  # 獲取每一個 date 在整個日期陣列中的索引 xi，以及對應的 spx 值 yi
    impt_tick.append(xi)
    yi = spx.asof(date)
    
    ax1.scatter(xi, yi, 80, color=r_hex)  # 用 scatter() 函式畫出一個圓點，標註事件在 spx 的位置
    # 在 annotate() 函式裡設定了事件，箭頭座標，事件列印的座標，箭頭性質，以及對齊屬性
    ax1.annotate(label, xy=(xi, yi+60), xytext=(xi, yi+300), 
                 arrowprops=dict(facecolor='black', headwidth=4, width=1, headlength=6), 
                 horizontalalignment='left', verticalalignment='top')

    
# 日期分成兩類：常規日期標籤 init_tick、關鍵事件日期　impt_tick
x_tick = init_tick+impt_tick
x_label = [x[i].strftime('%Y-%m-%d') for i in x_tick]
ax1.set_xticks(x_tick)
ax1.set_xticklabels(x_label, rotation=90)

for i, label in enumerate(ax1.get_xticklabels()):
    if i >= len(init_tick):
        label.set_color(r_hex)
        label.set_fontweight('bold')
    else:
        label.set_fontsize(9)

# 設定第二個座標軸
ax2 = ax1.twinx()
ax2.plot(y2, color=r_hex, linewidth=2, linestyle='-', label='VIX', alpha=0.3)
ax2.legend(loc='upper right', frameon=True)

plt.show()
```
